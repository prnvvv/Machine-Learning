```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, r2_score
```

```python
BreastCancerData = pd.read_csv('Breast_cancer_data.csv')
BreastCancerData.head(20)
```

```python
sns.heatmap(BreastCancerData.corr(), annot=True)
plt.show()
```

```python
sns.scatterplot(x = 'mean_radius', y = 'mean_texture', hue = 'diagnosis', data = BreastCancerData)
plt.show()
```

```python
sns.pairplot(BreastCancerData, hue = 'diagnosis')
plt.show()
```

```python
BreastCancerData.dropna(axis = 0, inplace = True)
```

```python
features = BreastCancerData.columns
features = features.drop(["diagnosis"])
X = BreastCancerData[features]
y = BreastCancerData["diagnosis"]
```

```python
scalar = StandardScaler()
X = scalar.fit_transform(X)
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = LogisticRegression()
model.fit(train_X, train_y)
prediction = model.predict(test_X)
```

```python
print(f"Accuracy score: {accuracy_score(test_y, prediction)}")
print(f"r2_score: {r2_score(test_y, prediction)}")
```

```python
model = RandomForestClassifier(random_state = 1)
model.fit(train_X, train_y)
prediction = model.predict(test_X)
```

```python
print(f"Accuracy score: {accuracy_score(test_y, prediction)}")
print(f"r2_score: {r2_score(test_y, prediction)}")
```
